```python
import xml.etree.ElementTree as ET

file_path = r"C:\Users\Maya Emad\Downloads\Graduation Project-Data Engineering\PostHistory.xml" 
# Read XML in an efficient way
context = ET.iterparse(file_path, events=("start", "end"))
_, root = next(context)  # Get root element

data = []
for event, elem in context:
    if event == "end" and elem.tag == "row":  # Adjust this based on your XML structure
        data.append(elem.attrib)  # Extract attributes
        elem.clear()  # Free memory

# Convert to DataFrame
df_PostHistory = pd.DataFrame(data)

print(df_PostHistory.head())  # Preview the data
```

```python
import re
import json
from bs4 import BeautifulSoup

# Function to clean text
def clean_text(text):
    if isinstance(text, str):
        try:
            # Step 1: Parse JSON if text is a valid JSON string
            text = json.loads(text) if text.strip().startswith("{") else text
            if isinstance(text, dict):
                text = text.get("some_key", "")  # Adjust based on JSON structure
        except json.JSONDecodeError:
            pass  # If it's not JSON, continue cleaning

        # Step 2: Remove HTML/XML tags
        text = BeautifulSoup(str(text), "html.parser").get_text()

        # Step 3: Remove any extra formatting (e.g., special characters)
        text = re.sub(r"<.*?>", "", text)  # Remove any remaining <tags>
        text = text.strip()  # Trim spaces

    return text

# Apply the function to the 'Text' column
df_PostHistory["Text_Cleaned"] = df_PostHistory["Text"].apply(clean_text)

# Display cleaned text
print(df_PostHistory[["Text", "Text_Cleaned"]].head())
```
